```python
weather_data = {"Day":['01/01/2022','01/02/2022','01/03/2022','01/04/2022','01/05/2022'],
                'Temperature':[32,25,50,40,42],
                'Windspeed':[5,6,8,7,4],                                                  #creating DataFrame using a dictionary
               'Event':['rain','sunny','snow','sunny','rain']}
df = pd.DataFrame(weather_data)
df
```

```python
weather_data = [
    ('01/01/2022',32,5,'rain'),                            #creating DataFrame using tuple
    ('01/02/2022',25,6,'sunny'),                # whenever we create DataFrame using tuple we need to put all tuple in a single list[]
    ('01/03/2022',50,8,'snow'),
    ('01/04/2022',40,7,'sunny'),
    ('01/05/2022',42,4,'rain')
    ]                           # we need to give columns name whenever we are creating DataFrame by tuples
df = pd.DataFrame(weather_data, columns = ['Day','Temperature','Windspeed','Event'])
df
```

### fillna()  function 


```python
df2 = df.fillna(0)
df2     # filling all na values with 0 and assinging this data to a defferent DataFrame df2
```

```python
# now see all the na values has been filled by 0

# but what if we want to fill partucular column's missing data with different different values, 
# in that condition we can pass dic in fillna ,, see bellow example 
```

```python
df2 = df.fillna({"Dry T.": 0,
                "Wet T.": 10,
                "Hum.": 50,
                 "Min. T.": 0})    
#we have passed dic for particular columns and have given to data to fill in that column
df2
```

### fillna(method ="ffill")   fuction 

```python
# this method carry forward the previous data to the na cell ,, 
```

```python
df2 = df.fillna(method = "ffill")
df2  # see bellow example how this method has carry forward the previous data to na cell 
```

### fillna(method ="bfill")   fuction 

```python
# instead of carry forward if you wanna back forward you can use this function 
```

```python
df2 = df.fillna(method="bfill")  # this gonna do back forward data to the empty cell 
df2
```

### axis = "column" 

```python
# it will copy data vertically from columns
```

```python
df2 = df.fillna(method = "ffill" ,axis ="columns")
df2    # this has copy data from right to left 
```

```python
df2 = df.fillna(method = "bfill" ,axis ="columns")
df2    # this has copy data from left to right 
```

### fillna , ffill & bfill limit 

```python
# if many cells are na in a series then we can apply limit function that up to how many empty cells we want to forward the data
```

```python
df2 = df.fillna(method = "ffill", limit = 2)
df2 # it will forward only upto two cells
```

### merge function

### Pivot table basic

```python
df.pivot(index='day',columns='city',values='windspeed')  # with value windspeed 
df 
```
